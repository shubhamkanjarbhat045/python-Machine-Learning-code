#Control Flow

##Conditional Statements

###Basic if-else

```python
def get_grade_letter(score):
  if score>=90:
    return "A"
  elif score>=80:
    return "B"
  elif score>=70:
    return "C"
  elif score>=60:
    return "D"
  else:
    return "F"

```

##Test The Function

```python
test_scores=[95,83,72,65,58]
for score in test_scores:
  grade=get_grade_letter(score)
  print(f"Score{score}:Grade{grade}")
```

##Nested Conditions

```python
def get_shipping_cost(weight, distance):
  base_cost = 5.0

  if weight <= 1:
    weight_cost = 2.0
  elif weight <= 5:
    weight_cost = 5.0
  else:
    weight_cost = 10.0

  if distance <= 100:
    distance_cost = 1.0
  elif distance <= 500:
    distance_cost = 5.0
  else:
    distance_cost = 10.0

  return base_cost + weight_cost + distance_cost
```

##Test shipping cost

```python
print(f"Light,close:${get_shipping_cost(0.5,50)}")
print(f"Heavy,far:${get_shipping_cost(10,800)}")
```

#Loops-While loop

##while loop for user input simulation

```python
def sim_guessing_game():
    target=7
    guess=0
    attempts=0
    print("Guess the number between 1 and 10!")
    while guess!=target:
     import random
     guess=random.randint(1,10)
     attempts+=1
     if guess<target:
        print(f"guess{attempts}:{guess}-Too Low")
     elif guess>target:
        print(f"guess{attempts}:{guess}-Too High")
     else:
        print(f"guess{attempts}:{guess}-Correct")
     if attempts>10:
      print("Max attempts reached")
      break
    print(f"Game finished in {attempts}attempts")
sim_guessing_game()
```

##While Loop for Processing

```python
def process_queue():
  queue=["task1","task2","task3","task4"]
  while queue:
    current_task=queue.pop(0)
    print(f"processing:{current_task}")
    print(f"Remaining tasks:{len(queue)}")
  print("All tasks completed")
process_queue()
```

#Advance Function concepts

##functioning returning multiple values

```python
def analyze_text(text):
  words=text.split()
  word_count=len(words)
  char_count=len(text)
  char_count_no_spaces=len(text.replace(" ",""))
  return word_count,char_count,char_count_no_spaces
```

##Test with sample Text

```python
sample_text="python is powerful programming language"
words,chars,chars_no_spaces=analyze_text(sample_text)
print(f"text:'{sample_text}'")
print(f"word_count:{words}")
print(f"char_count:{chars}")
print(f"char_count_no_spaces:{chars_no_spaces}")
```

##variable lenth arguments

```python
def calculate_average(*numbers):
  if not numbers:
    return 0
  return sum(numbers)/len(numbers)
```

##Test with diff numbers of arguments

```python
avg1=calculate_average(10,20,30)
avg2=calculate_average(5,10,15,20,25)
avg3=calculate_average(100)
print(f"Average1:{avg1}")
print(f"Average2:{avg2}")
print(f"Average3:{avg3}")
```

#Lambda Creations

##lamba Functions(anonyms functions)

###regular functi()

```python
def square(x):
  return x**2
```

###Lambda Equivalent

```python
square_lambda=lambda x:x**2
print(f"regular:{square(5)}")
print(f"lambda:{square_lambda(5)}")
```

###lamba with multiple arguments

```python
multiply=lambda x,y:x*y
print(multiply(5,3))
```

###Lambda with built-in ()'S

```python
numbers=[1,2,3,4,5]
squared=list(map(lambda x:x**2,numbers))
print(f"squared:{squared}")
```

##Filtering with lambda

```python
ages=[16,21,35,12,45,18,67]
adults=list(filter(lambda age:age>=18,ages))
print(f"adults:{adults}")
```

##sorting with Lambda

```python
people=[
    {"name":"Alice","age":25},
    {"name":"Bob","age":30},
    {"name":"Eve","age":20}
]
sorted_people=sorted(people,key=lambda person:person["age"])
print(f"sorted_people by Age:")
for person in sorted_people:
  print(f"{person['name']}:{person['age']}")
```
